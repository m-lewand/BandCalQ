```python
import numpy as np
from qiskit.providers.basicaer import QasmSimulatorPy
#from bandcalq import BandCalQ
import bandcalq
from qiskit.algorithms.optimizers import Optimizer, Minimizer, SPSA
from importlib import reload
reload(bandcalq)
```

```python
#Minimal amount of parameters
orbitals = 2
t = np.array([[ 0.04    ,  -0.08],
             [ -0.08    ,   0.06 ]])
lattice_parameter = 2
backend = QasmSimulatorPy()

#Defining problem using BandCalQ class
problem = bandcalq.BandCalQ(orbital_number=orbitals, lattice_constant=lattice_parameter, hopping_matrix=t, backend=backend, optimizer=SPSA(maxiter=2))

#Computing bands, in this example inside first Brillouin zone
#Theoretical points can be calculated in compute_band_structure method or later in plot_band_structure
problem.compute_band_structure(momentum_min=np.pi/lattice_parameter, momentum_max=np.pi/lattice_parameter, momentum_points_amount=1, theoretical_points=False)
```

```python
problem.plot_band_structure(theoretical_points=False, save_png=True, png_name='test')
```
